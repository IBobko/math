```python
import numpy as np
import matplotlib.pyplot as plt

def f1(n, x, index):
    """
    n -  
    x - ,      
    index -   .
    """
    return n + x * index

def period(start, increase, iteration_count,divide):
    x = list()
    y = list()
    sum = 0
    for i in range(0,iteration_count):
        x.append(i * divide + divide)
        sum = sum + f1(start,increase,i)
        y.append(sum)
       # print("i = {} x={} y= {}".format(i, i * divide + divide, sum))
    return x,y



def gperiod(start, increase, iteration_count,divide):
    return period(start/divide,8000/divide,3*divide,1/divide)

fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)
for i in range(1,6):
    x_value, y_value = gperiod(40000,8000,3,i)
    ax.plot(x_value, y_value)    
#ax.set_xticks([0,1,2,3,4])
#ax.set_yticks(range(0,200000,5000))
plt.grid()
plt.show()
```
